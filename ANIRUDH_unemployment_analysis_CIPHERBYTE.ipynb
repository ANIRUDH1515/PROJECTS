```python
plt.figure(figsize=(12, 8))

sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=unemployment_india_df)

plt.title('Trend of Unemployment Rate in India')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Region', y='Estimated Unemployment Rate (%)', data=unemployment_india_df)
plt.title('Box Plot of Unemployment Rate by Region')
plt.xlabel('Region')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.show()

```
